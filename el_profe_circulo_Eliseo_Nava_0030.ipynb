## formua para calcular perimetro
$P=2*\pi*r$

## Datos de entrada

```python
# calcula area y perimetro de un circulo
# Variables A, P, r,
# formulas A = pi*r^ y P = 2 * pi * r
# Calcular el valor de r
import math
print("Calcular area y perimetro de un circulo")
r = float(input("Ingrese el valor de r: "))
A =math.pi*r**2
P = 2*math.pi*r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Cepeda Darrell NL = 10")
```
